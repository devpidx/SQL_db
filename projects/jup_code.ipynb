```python
 # SQLAlchemy
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
import pandas as pd

# Path to sqlite
database_path = "product_codes.db"
```

```python
# Create an engine that can talk to the database
engine = create_engine(f"sqlite:///{database_path}")
```

```python
 # Query All Records in the the Database
data = engine.execute("SELECT * FROM product_codes_tb")

for record in data:
    print(record)
    # note how you can use . notation to extract specific columns
    print(record.code)
```

```python
names = engine.execute("SELECT * FROM product_codes_tb WHERE product_definition LIKE '%%ethanol%%';")

for record in names:
    print(record)
    # note how you can use . notation to extract specific columns
    print(record.code)
```
